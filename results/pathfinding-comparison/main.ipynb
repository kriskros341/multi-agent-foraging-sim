```python
import pandas as pd

import pandas as pd
import json

# Load the full JSON into a dictionary
with open('greedy_results.json', 'r') as f:
    data = json.load(f)

# Navigate to the specific key (e.g., data -> 'results' -> 'users')
greedy = pd.DataFrame(data['summaries'])
greedy.describe()
```

```python
# Percentage of runs that finished successfully
successful_astar_runs = astar[(astar['endingTypeCounts_RESOURCE'] == 0) & (astar['endingTypeCounts_SCOUTED_RESOURCE'] == 0)]
successful_astar_runs.shape[0] / astar.shape[0] * 100
```

```python
successful_greedy_runs = greedy[(greedy['endingTypeCounts_RESOURCE'] == 0) & (greedy['endingTypeCounts_SCOUTED_RESOURCE'] == 0)]
successful_greedy_runs.shape[0] / greedy.shape[0] * 100
```

```python
first_successful_astar = successful_astar_runs.iloc[0]
print(first_successful_astar['final_environment_snapshot'])
```

```python
astar_beat_greedy = successful_astar_runs[successful_astar_runs['iter'] < successful_greedy_runs['iter'].min()]
astar_beat_greedy
```

```python
greedy_beat_astar = successful_greedy_runs[successful_greedy_runs['iter'] < successful_astar_runs['iter'].min()]
greedy_beat_astar
astar[astar['seed'] == greedy_beat_astar.iloc[0]['seed']]
```

```python
greedy.plot.hist(column='iter', bins=30)
astar.plot.hist(column='iter', bins=30)
```
